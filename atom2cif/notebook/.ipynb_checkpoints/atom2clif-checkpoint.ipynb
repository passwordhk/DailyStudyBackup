```python
sheet = pd.read_csv(r'/Users/huangkai/Documents/DailyStudyBackup/atom2cif/out2pandas.txt',sep=' ',engine='python')
sheet
```

```python
df=sheet.dropna(axis=1) # 
df
```

```python
type(sheet['type'])
```

```python
df['type']=df['type'].apply(lambda x: 'C' if x == 1 else ('H' if x == 2 else('O' if x == 3 else('Ca' if x==4 else 'NaN'))))
```

```python
df.head(15)
```

```python
atom_count_C = 0
atom_count_H = 0
atom_count_O = 0
atom_count_Ca = 0
arr = []
for index,type in zip(df['id'],df['type']):
#     print(index,type)
    if type == 'C':
        atom_count_C += 1
        index = 'C{}'.format(str(atom_count_C))
        arr.append(index)
    elif type == 'H':
        atom_count_H += 1
        index = 'H{}'.format(str(atom_count_H))
        arr.append(index)
    elif type == 'O':
        atom_count_O += 1
        index = 'O{}'.format(str(atom_count_O))
        arr.append(index)
    elif type == 'Ca':
        atom_count_Ca += 1
        index = 'Ca{}'.format(str(atom_count_Ca))
        arr.append(index)
# print(arr)        
series_1=pd.Series(arr)
df['id'] = series_1
```
