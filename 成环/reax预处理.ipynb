```python
import re

def istype_convert(s):
    a = None
    if s in O_index:
        a = f'O{s}'
    elif s in C_index:
        a = f'C{s}'
    elif s in H_index:
        a = f'H{s}'
    return a

def convert_index(li):
    for i,j in enumerate(li):
        if j=='0':
            b = [0] + [k for k in range(3,i)]
            return b

def get_count_rear(li):
    for index,value in enumerate(li):
        if value == '0':
            return index-3
        

def del_H(li):
    sc = []
    for i in li:
        if i.startswith('H'):
            sc.append(i)
    no_H  = [k for k in li if k not in sc]
    return no_H
        
f = open('1660k.data','r')
all_li = f.readlines()
new = all_li[16:]
raw_content0 = []
for i in new:
    raw_content0.append(i.strip().split())
for i in raw_content0:
    if i[1] == '1':
        i[0] = f'C{i[0]}'
    elif i[1] == '2':
        i[0] = f'H{i[0]}'
    elif i[1] == '3':
        i[0] = f'O{i[0]}'
C = []
H = []
O = []
for i in raw_content0:
    if i[0].startswith('C'):
        C.append(i[0])
    elif i[0].startswith('H'):
        H.append(i[0])
    elif i[0].startswith('O'):
        O.append(i[0])
# 获取原子固定编号
O_index = [i[1:] for i in O]
C_index = [i[1:] for i in C]
H_index = [i[1:] for i in H]
f.close()




f2=open('bonds.reax')
con=f2.readlines()[7:]
a=[]
for i in con:
    a.append(i.strip().split())
    
for i in range(len(a)):
    for j in convert_index(a[i]):
        a[i][j] = istype_convert(a[i][j])

f_temp = open('temp.reax','w')
C_O = []
for i in a:
    f_temp.write('\t'.join(i)+'\n')
f_temp.close()
f2.close()

with open('temp.reax') as f:
    content_raw = f.readlines()
    content = []
    for i in content_raw:
        content.append(i.strip().split())
        
del_index = []
for i in content:
#     print(i)
    if i[0].startswith('H') or i[0].startswith('O'):
        del_index.append(i)
        
tep_content = [i for i in content if i not in del_index]

goal = []
for i in tep_content:
#     print(i)
    temp_li = []
    temp_li.append(i[0])
    a = i[3:3+get_count_rear(i)]
    for k in a:
        temp_li.append(k)
    goal.append(temp_li)
print(len(goal))
out = []
for i in goal:
    out.append(del_H(i))
for i in out:
    if len(i) == 3 and i[2].startswith('O'):
        out.remove(i)
f_out = open('final_data.reax','w')
for i in out:
    f_out.write(str(i)+'\n')

hot_pot = dict()
for i in out:
    hot_pot[i[0]] = i[1:]
```
