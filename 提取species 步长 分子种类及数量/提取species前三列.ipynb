```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
f = open('species.out')
new_f = open('Time_Mole_Spec.out','w')
raw = f.readlines()
new = []
new.append(raw[0].strip('# ').split()[0:3])
for i in raw[1:]:
    if not i.startswith('#'):
        new.append(i.strip().split()[0:3])
for i in new[1:]:
    i[0] = str(float(i[0])/10000)
for i in new:
    new_f.write('\t'.join(i)+'\n')
new_f.close()
f.close()
```

```python
df = pd.read_csv('Time_Mole_Spec.out',sep='\t',engine='python',index_col=None)
f,ax = plt.subplots(figsize=(15,6))
g1=sns.relplot(x='Timestep',y='No_Moles',kind='line',data=df,ax=ax)
g2=sns.relplot(x='Timestep',y='No_Specs',kind='line',data=df,ax=ax)
g1.fig.autofmt_xdate()
g2.fig.autofmt_xdate()
plt.close(2)
plt.close(3)
```
